```
import statsmodels.api as sm

predictor_variables = [
'Qué tanto apoyo recibes de tu familia para estudiar temas relacionados con las matemáticas o la ciencia',
'Tus maestros te han animado a explorar temas relacionados con la ciencia o la tecnología',
'Asignaturas_Tecnología',
'Asignaturas_Matematica',
'Qué opinan tus amigos sobre el estudio de disciplinas relacionadas con la tecnología o la ciencia',
'Sector_de_interes_futuro_Tecnología',

]

target_variable = 'Cuál es la probabilidad de que selecciones una carrera relacionada con Ciencia, Tecnología, Ingeniería, Arte o Matemáticas (STEAM)'

X = df_numeric[predictor_variables]
Y = df_numeric[target_variable]

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

model_summary = model.summary()
model_summary
```

```
for predictor in predictor_variables:
    sns.lmplot(x=predictor, y=target_variable, data=df_numeric, line_kws={'color': 'red'})
    plt.title(f'Regresión Lineal entre {predictor} y {target_variable}')
    plt.xlabel(predictor)
    plt.ylabel(target_variable)
    plt.show()
```

```
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

#Residuos d enuestro modelo
residuos = model.resid

# Shapiro-Wilk
stat, p_value = stats.shapiro(residuos)
print(f'Shapiro-Wilk Test: p-value = {p_value}')

# Gráfico 
stats.probplot(residuos, dist="norm", plot=plt)
plt.show()
```

```
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
import matplotlib.pyplot as plt

# --- 3. MULTICOLINEALIDAD (Factor de Inflación de Varianza - VIF) ---
# Calcula VIF para cada variable independiente
vif_data = pd.DataFrame()
vif_data["Variable"] = model.model.exog_names  # Nombres de las variables
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

print("\nFactor de Inflación de Varianza (VIF):")
print(vif_data)
```
